```python
import json
import numpy as np
import matplotlib.pyplot as plt
with open('abuse_report.json') as f:
    df = json.load(f)

pd.DataFrame(df)
import pandas as pd
data=pd.DataFrame(df)
print(data.head())
```

```python
data_dict={}
for key, value in data.items():
    print(f"Key: {key}/ Value:{len(value)}")
    data_dict[key]=value
#print(data_dict)
```

```python
print(data_dict)
```

```python
df=pd.DataFrame(data_dict)
df.T.head()
```

```python
df.T['sentiment']
```

```python
def extract_sentiment_data(sentiment_str):
    output={}
    sentiment_dict=dict(sentiment_str)
    for key, value in sentiment_dict.items():
        if  'sentiment' not in value:
            continue
        output[key]=value["sentiment"].lower()
    return output
```

```python
df.T['sentiment'].apply(lambda x: extract_sentiment_data(x))
```

```python
new_data=df.T['sentiment'].apply(lambda x: extract_sentiment_data(x))
```

```python
for key, value in new_data.items():
    for k, v in value.items():
        if k == "roberta":
            if v== "label_0":
                value[k]="negative"
            elif v== "label_1":
                value[k]="neutral"
            else:
                value[k]="positive"
            # 
            # "
```

```python
new_df=pd.DataFrame(new_data.tolist())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your DataFrame is already loaded as new_df
models = ["bert", "vader", "roberta", "textblob"]

# Count sentiment occurrences per model
sentiment_counts = {model: new_df[model].value_counts() for model in models}
sentiment_df = pd.DataFrame(sentiment_counts).fillna(0)

# Plotting
sentiment_df.plot(kind="bar")
plt.ylabel("Count")
plt.xlabel("Sentiment")
plt.title("Sentiment Distribution per Model")
plt.xticks(rotation=0)
plt.show()
```
